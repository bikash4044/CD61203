```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model


URL = "https://raw.githubusercontent.com/GLambard/Molecules_Dataset_Collection/master/latest/ESOL_delaney-processed.csv"
df = pd.read_csv(URL, index_col=0)

df.rename(columns={"measured log solubility in mols per litre": "SOL"}, inplace=True)

reg = linear_model.LinearRegression()
reg.fit(df[["Minimum Degree","Molecular Weight","Number of H-Bond Donors","Number of Rings","Number of Rotatable Bonds","Polar Surface Area"]],df.SOL)
reg.coef_
# reg.intercept_
```
